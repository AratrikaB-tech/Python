Activity #1: Reading an original Sequence

```python
from Bio.Seq import Seq
seq = Seq("ATGCGTATGC")
print ("seq", seq)
```

Activity #2: GC Content Calculation

```python
from Bio.SeqUtils import gc_fraction
Seq = "ATGCGTGCGTGCGTGCGTGCGTGCGCGCGTGTTA"
print("GC content:", gc_fraction(Seq))
```

Activity #3: Transcription -- DNA -> RNA

```python
from Bio.Seq import Seq
dna = Seq("ATGCTGAGTTCGGCGTGCTGGTTCCAATGATATC")
rna = dna.transcribe()
print("RNA:", rna)
```

Activity #4: Direct Translation --- DNA -> Protein

```python
from Bio.Seq import Seq
dna = Seq("ATGCCGTAGGCGGCCGCTTAAATATGGCCGTAAGCTGCTATTTGTGC")
protein = dna.translate()
print("Protein:", protein)
```

Activity #5: Reverse Complement --- Using a Diff. Sequence

```python
from Bio.Seq import Seq
dna = Seq("ATCGGTGCGTCGGAAATGCTAATGCCTGTAGC")
print("Reverse Complement:", dna.reverse_complement())
```

Activity #6: Counting Nucleotides in dna/rna

```python
from Bio.Seq import Seq
dna = Seq("ATGCCGTATGCGTGCAATTGGGCCCA")
for base in "ATGC":
  print (base, "=", dna.count(base))
```
